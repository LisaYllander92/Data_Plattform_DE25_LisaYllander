## Profit per country

```python
import duckdb

duckdb.query(
    """
    SELECT country, SUM(profit) as profit
    FROM df
    GROUP BY country
    ORDER BY profit desc"""
)
```

```python
# Convert to json
duckdb.query(
    """
    SELECT country, SUM(profit) as profit
    FROM df
    GROUP BY country
    ORDER BY profit desc"""
).df().to_json()
```
